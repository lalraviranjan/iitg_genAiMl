# 5 number Summary And Box Plot

### Minimum,MAximum,Median,Q1,Q3,IQR

```python
import numpy as np
```

#### First we will see list of marks without any outliers using boxplot

```python
lst_marks=[45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74]
minimum,q1,median,q3,maximum = np.quantile(lst_marks, [0, 0.25, 0.50, 0.75, 1.0])
```

```python
print(minimum,q1,median,q3,maximum)
```

```python
IQR = q3 - q1
print(IQR)
```

```python
lower_fence = q1- 1.5 * IQR
higher_fence= q3 + 1.5 * IQR
```

```python
print(lower_fence, higher_fence)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# without any outliers as higher fence is 141.5 and lower fence is 1.5 and we dont have these numbers in list
sns.boxplot(lst_marks)
plt.title("Boxplot of Marks Distribution - 5 number Summary without Outliers", fontsize=16)
plt.show()
```

#### List of Marks with Outliers and Boxplot indicating outliers

```python
lst_marks=[-120,-140,-2,45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74,190, 210,200]
```

```python
minimum,q1,median,q3,maximum = np.quantile(lst_marks, [0, 0.25, 0.50, 0.75, 1.0])
print(minimum,q1,median,q3,maximum)
```

```python
IQR = q3 - q1
print(IQR)
```

```python
lower_fence = q1- 1.5 * IQR
higher_fence= q3 + 1.5 * IQR
print(lower_fence, higher_fence)
```

```python
# with outliers as higher fence is 141.5 and lower fence is 1.5 and we dont have these numbers in list
sns.boxplot(lst_marks)
plt.title("Boxplot of Marks Distribution - 5 number Summary with Outliers", fontsize=16)
plt.show()
```
