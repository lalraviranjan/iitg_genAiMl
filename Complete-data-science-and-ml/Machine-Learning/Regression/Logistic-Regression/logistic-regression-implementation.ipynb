## Hyperparameter Tuning using GridSearchCV
#### Check Sklearn Logistic Regression Paremeters - https://scikit-learn.org/1.5/modules/generated/sklearn.linear_model.LogisticRegression.html

### What is GridSearchCV?

`GridSearchCV` is a hyperparameter tuning method in machine learning provided by the `scikit-learn` library. It is used to find the optimal combination of hyperparameters for a given model. The "CV" stands for **cross-validation**, meaning it evaluates the performance of different hyperparameter combinations using a cross-validation strategy.

---

### When Do We Use GridSearchCV?

We use `GridSearchCV` when:
1. **Optimizing Hyperparameters**:
   To find the best combination of parameters that maximizes model performance. For example, finding the best value for `C` in Logistic Regression or the depth of a Decision Tree.
   
2. **Automating the Search**:
   Instead of manually testing various combinations, `GridSearchCV` performs the process automatically.

3. **Evaluating Multiple Hyperparameters Simultaneously**:
   It can evaluate combinations of multiple hyperparameters, e.g., `C` and `penalty` in Logistic Regression.

4. **Preventing Overfitting**:
   Cross-validation ensures the model performs well on unseen data by dividing the dataset into training and validation splits multiple times.

---
### **Automatic Handling of Invalid Combinations**:
   - The underlying library may skip combinations that are not compatible, especially with penalties and solvers in `LogisticRegression`. For example:
     - `lbfgs` does not support `penalty='l1'`, but the grid search may avoid running that specific pairing.


````python
#Now we will use the paramters avaialble in LogisticRegression
model = LogisticRegression()
penalty=['l1', 'l2', 'elasticnet']
c_value = [100, 10, 1.0, 0.1, 0.01]
solver = ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
````

````python
params = dict(penalty=penalty, C=c_value, solver=solver)
````

````python
## GridSearchCV
from sklearn.model_selection import GridSearchCV, StratifiedKFold
cv = StratifiedKFold()
grid = GridSearchCV(estimator=model, param_grid=params, scoring='accuracy', cv=cv, n_jobs=-1)
````

````python
# (.fit) Train the model/ perform cross validation and identify best params
grid.fit(X_train, Y_train)
````

#### What is `grid.fit(X_train, Y_train)`?

The `.fit()` method in `GridSearchCV` is used to:
1. **Train the Model**: It trains the underlying model (e.g., `LogisticRegression`) for every combination of hyperparameters provided in the parameter grid.
2. **Perform Cross-Validation**: It performs the specified cross-validation (e.g., StratifiedKFold) on the training data for each hyperparameter combination.
3. **Identify the Best Parameters**: It evaluates each model's performance and identifies the hyperparameters that maximize the specified scoring metric (e.g., accuracy).

### Explanation of `Penalty`,  `c_value` and `solver` Parameters in Logistic Regression:

### **The Role of Penalty**
1. **Penalty** defines the **type** of regularization to be applied.
   - **L1 Regularization**: Adds the sum of absolute coefficients to the loss function, leading to sparsity (some coefficients become zero).
   - **L2 Regularization**: Adds the sum of squared coefficients to the loss function, leading to smooth weight values.
   - **ElasticNet**: Combines both L1 and L2 regularization, controlled by the **`l1_ratio`**.

2. The penalty specifies **what kind of constraint** the model will be subjected to.

---

### **The Role of `C`**
1. **`C` controls the strength of the regularization**, irrespective of the penalty type.
   - **Larger `C`**: Less regularization (focuses more on fitting the training data).
   - **Smaller `C`**: Stronger regularization (simplifies the model and reduces overfitting).

2. `C` is **tunable**, allowing us to find the best balance between overfitting and underfitting based on the dataset. 
   
3. Without the **`C` value**, you wouldn't have control over the intensity of regularization.

 **Values**: 
   - `[100, 10, 1.0, 0.1, 0.01]`
   - **100 and 10**: Weak regularization (complex model).
   - **1.0**: Moderate regularization.
   - **0.1 and 0.01**: Strong regularization (simpler model).

   **Purpose**: To find the optimal trade-off between underfitting and overfitting through hyperparameter tuning.

---
### Why Do We Need This?
- Regularization (L1/L2/ElasticNet) keeps the model simple and avoids overfitting.  
- `C` gives you the **flexibility to adjust the intensity** based on your dataset.

---
#### **The Role of `solver` (Optimization Algorithm)**:
   - Defines the **algorithm** used to find the weights/parameters of Logistic Regression by minimizing the cost function.
   - Different solvers are suitable for different scenarios:

   **Purpose**: By using different solvers, we identify the one best suited for our data and regularization choice.

#### **1. `n_clusters_per_class`**
- This parameter specifies the number of clusters of points to generate for each class.
- **What does it mean?**
  If `n_clusters_per_class=1`, all the points of a particular class will form one cluster. For example:
  - Class 0 points will form a single cluster in feature space.
  - Class 1 points will also form a single cluster.

  If `n_clusters_per_class > 1`, points for each class will form multiple, distinct clusters.

- **Why is it important?**
  - Increasing `n_clusters_per_class` makes the classification problem more complex because classes are now divided into subgroups (clusters) that overlap with other class clusters.
  - A small value (e.g., 1) creates simpler, easier-to-separate datasets, which is useful for visualizations or testing basic models.

---

#### **2. `weights`**
- This parameter specifies the proportions of samples to generate for each class.
- **What does it mean?**
  - `weights=[0.99]` means 99% of the data points will belong to the **majority class** (Class 0), and the remaining 1% will belong to the **minority class** (Class 1).
  - The weights list must sum to 1. If there are only two classes, you only need one weight (the complement will be assigned to the second class).

- **Why is it important?**
  - This parameter simulates **class imbalance**, which occurs in many real-world datasets (e.g., fraud detection, disease detection, rare event classification).
  - A highly imbalanced dataset helps you test a model's ability to handle such scenarios, as most algorithms tend to favor the majority class.

## Hyperparamter Tuning withGridSearchCV for imbalanced datasets

````python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()

#Now we will use the paramters avaialble in LogisticRegression
model = LogisticRegression()
penalty=['l1', 'l2', 'elasticnet']
c_value = [100, 10, 1.0, 0.1, 0.01]
solver = ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
class_weight=[{0:w,1:y} for w in [1,10,50,100] for y in [1,10,50,100]]

params = dict(penalty=penalty, C=c_value, solver=solver, class_weight=class_weight)
````

#### **Explanation of `class_weight`:**
1. **General Idea:**
   - Logistic Regression treats all misclassifications equally unless we introduce a **class weight**.
   - By adjusting the importance (weight) of each class, we tell the model to focus more on the minority class or balance the influence of each class during training.

2. **Weights in the Code:**
   ```python
   class_weight = [{0: w, 1: y} for w in [1, 10, 50, 100] for y in [1, 10, 50, 100]]
   ```
   - The weights `{0: w, 1: y}` allow us to assign different importance levels to classes `0` and `1`.
   - For example:
     - `{0: 1, 1: 1}` means both classes have equal importance.
     - `{0: 1, 1: 10}` makes class `1` ten times more important than class `0`.
     - `{0: 100, 1: 50}` makes class `0` twice as important as class `1`.

---

### **Purpose of the Code:**
- The goal of adding `class_weight` to the hyperparameter grid is to identify **which weight combination performs best** in terms of accuracy.
- In imbalanced datasets, models often underperform on the minority class;

# Logistic Regression With **`ROC curve`** and **`ROC AUC`** Score

#### **ROC Curve (Receiver Operating Characteristic Curve)**

The **ROC curve** is a graphical representation of the diagnostic ability of a binary classifier as its discrimination **`threshold`** is varied. It is used to evaluate the performance of classification models by plotting the **true positive rate (TPR)** against the **false positive rate (FPR)** at different threshold settings.

#### **ROC AUC Score (Area Under the ROC Curve)**

The **AUC (Area Under the Curve)** is a metric that quantifies the overall performance of the classifier across all possible thresholds. It represents the area under the ROC curve, providing a single number to evaluate the modelâ€™s ability to discriminate between the positive and negative classes.

#### **AUC Explanation**:
- **AUC = 1**: A perfect model that can perfectly distinguish between all positive and negative cases.
- **AUC = 0.5**: A model with no discrimination ability (random guessing).
- **AUC < 0.5**: Indicates that the model performs worse than random guessing (can be interpreted as a bad model, and potentially needs reworking).

#### **Why AUC is Useful**:
- AUC is independent of the decision threshold, meaning that it provides an aggregate measure of the modelâ€™s ability to distinguish between classes, without considering how you might set the cut-off point (the threshold between classifying an observation as positive or negative).
- A model with a high AUC generally means it performs well in identifying positive and negative instances correctly.

#### **In Summary**:

1. **ROC Curve**:
   - Plots **True Positive Rate (TPR)** vs. **False Positive Rate (FPR)** at various thresholds.
   - Helps to understand the trade-off between detecting positive cases and avoiding false positives.
   
2. **ROC AUC Score**:
   - A numerical summary of the ROC curve.
   - Tells how well a classifier distinguishes between positive and negative cases, with values between 0 and 1.

### We trained two models: one using a simple Logistic Regression and another using GridSearchCV for hyperparameter tuning. Then, we plotted the ROC curves for both models to compare their performance. The model with a higher Area Under the Curve (AUC) typically predicts better.

### When analyzing the ROC curve, our goal is to increase the True Positive Rate (TPR), but we must also be cautious as increasing TPR can lead to an increase in the False Positive Rate (FPR).

### Next, we plotted the threshold values at different points on the ROC curve to see where the True Positives were high. By examining these thresholds, we can work with domain experts to choose the most suitable threshold for the model's specific needs.

````python
# Plotting ROC curves
plt.figure(figsize=(10, 6))
plt.plot(fpr_lr, tpr_lr, color='blue', label=f'Logistic Regression (AUC = {roc_auc_score(Y_test, y_pred_logistic):.2f})')
plt.plot(fpr_grid, tpr_grid, color='red', label=f'GridSearchCV Model (AUC = {roc_auc_score(Y_test, y_pred_grid):.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.title('ROC Curve Comparison')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
````

````python
# Plotting Thresholds on the same plot
plt.scatter(fpr_lr, tpr_lr, marker='x', color='blue', label="Logistic Thresholds")
for i in range(0, len(thresholds_lr), int(len(thresholds_lr)/5)):  # Adjust for showing few thresholds
    plt.text(fpr_lr[i], tpr_lr[i], f'{thresholds_lr[i]:.2f}', fontsize=10, color='blue', ha='left', va='bottom')

plt.scatter(fpr_grid, tpr_grid, marker='x', color='red', label="GridSearch Thresholds")
for i in range(0, len(thresholds_grid), int(len(thresholds_grid)/5)):  # Adjust for showing few thresholds
    plt.text(fpr_grid[i], tpr_grid[i], f'{thresholds_grid[i]:.2f}', fontsize=10, color='red', ha='left', va='bottom')

# Adding diagonal line (random classifier)
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')

# Adding plot details and zooming in
plt.xlim([0, 1])  # Zoom into the range for X-axis
plt.ylim([0, 1])  # Zoom into the range for Y-axis

plt.title('ROC Curve Comparison with Thresholds')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.grid(True)

# Show plot
plt.show()
````
