```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X = 6 * np.random.rand(100, 1) - 3
Y = 0.5 * x**2 + 1.5*X + 5 + np.random.randn(100,1)
plt.scatter(x=X, y=Y, color='red')
```

### We will solve this problem first with Linear regression and then will use Polynomial regression so that we can mismize the error while getting the best fit line.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
regression_1 = LinearRegression()
```

```python
regression_1.fit(X_train, Y_train)
```

### We can see using Linear Regression we get some kind os error and score woth r2.
### We can also see how the fbest fit line is there but the error are more scattered with respect to the best bit line.

```python
from sklearn.metrics import r2_score
y_pred = regression_1.predict(X_test)
score = r2_score(Y_test, y_pred)
print(score)
```

```python
plt.plot(X_train, regression_1.predict(X_train), color="r")
plt.scatter(X_train, Y_train)
```

### Now lets solve this problem using polynomial regression


```python
#lets apply polynomial transformation
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=2, include_bias=True)
```

```python
poly
```

```python
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
```

```python
X_train_poly
```

```python
X_test_poly
```

```python
from sklearn.metrics import r2_score
regression= LinearRegression()
regression.fit(X_train_poly, Y_train)
y_pred = regression.predict(X_test_poly)
score=r2_score(Y_test, y_pred)
print(score)
```

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
plt.scatter(X_train, regression.predict(X_train_poly))
plt.scatter(X_train,Y_train)
```

```python
poly = PolynomialFeatures(degree=3, include_bias=True)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
```

```python
X_train_poly
```

```python
regression= LinearRegression()
regression.fit(X_train_poly, Y_train)
y_pred = regression.predict(X_test_poly)
score=r2_score(Y_test, y_pred)
print(score)
```

```python
plt.scatter(X_train, regression.predict(X_train_poly))
plt.scatter(X_train,Y_train)
```

```python
X_new = np.linspace(-3,3,200).reshape(200,1)
X_new_poly = poly.transform(X_new)
X_new_poly
```

```python
y_new = regression.predict(X_new_poly)
plt.plot(X_new, y_new, "r-", linewidth=2, label="Predictions")
plt.plot(X_train, Y_train, "b.", label="Training Points")
plt.plot(X_test, Y_test, "g.", label="Testing Points")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
