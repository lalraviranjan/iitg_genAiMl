# Anomaly Detection - Isolation Forest

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("disease_detection_dataset.csv")
df.head()
```

```python
sns.scatterplot(x = df.iloc[:,0], y=df.iloc[:,1])
plt.show()
```

from sklearn.ensemble import IsolationForest

```python
clf = IsolationForest(contamination=0.2)
clf.fit(df)
predictions = clf.predict(df)
```

##### Wherever the values are 1 are the normal datapoints and wherever we are gettig as -1 are basically the outlier

```python
predictions
```

```python
index_outliers = np.where(predictions < 0)
index_outliers
```

```python
X= df.values
X.shape
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=df.iloc[:,0], y=df.iloc[:,1], label="Normal", color="blue")

# Plot outliers in red
# X[index_outliers, 0] → Extracts the first column values (Feature 1) of the outliers.
# X[index_outliers, 1] → Extracts the second column values (Feature 2) of the outliers
sns.scatterplot(x=X[index_outliers, 0], y=X[index_outliers, 1], label="Outliers", color="red")
```
