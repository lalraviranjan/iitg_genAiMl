```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
```

```python
# Generates a synthetic dataset using the "moons" shape  
X,Y = make_moons(n_samples=1000, noise=0.05)
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=X[:,0], y=X[:,1], hue=Y)
```

```python
# feature scaling : standard scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_scaled = scaler.fit_transform(X)
```

```python
# Initialize DBSCAN clustering algorithm  
# eps=0.3 -> Defines the radius within which neighboring points are considered part of the same cluster  
# metric='euclidean' -> Uses Euclidean distance to measure the closeness of points  
dbscan = DBSCAN(eps=0.3, metric='euclidean', n_jobs=-1)
```

```python
dbscan.fit(x_scaled)
```

```python
dbscan.labels_
```

```python
# Creates a scatter plot to visualize the clusters formed by DBSCAN  
plt.figure(figsize=(8,6))
plt.title("Scatter plot to visualize the clusters formed by DBSCAN ")
sns.scatterplot(x = x_scaled[:,0], y = x_scaled[:,1], hue=dbscan.labels_)
plt.show()
```
