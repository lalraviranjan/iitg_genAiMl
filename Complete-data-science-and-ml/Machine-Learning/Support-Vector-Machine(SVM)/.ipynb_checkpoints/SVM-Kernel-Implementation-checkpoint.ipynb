```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set a random seed for reproducibility
np.random.seed(42)

# Generate x1 and x2 features
x1 = np.random.randn(10000)
x2 = np.random.randn(10000)

# Generate the target y with a polynomial relationship
y = (x1 ** 2 + x2 ** 2 > 1.5).astype(int)

# Create a DataFrame
data = pd.DataFrame({'x1': x1, 'x2': x2, 'y': y})
```

```python
X = data.iloc[:,:2]
X.head()
```

```python
Y = data.iloc[:, 2]
Y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30, random_state=42)
```

```python
# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(data['x1'], data['x2'], c=data['y'], cmap='cool', edgecolor='k', alpha=0.7)
plt.title('Scatter Plot of Polynomial-like Data', fontsize=16)
plt.xlabel('x1')
plt.ylabel('x2')
plt.colorbar(label='Class (y)')
plt.grid(True)
plt.show()
```

```python
# Create the derived columns: x1^2, x2^2, and x1*x2
x_square = data['x1'] ** 2
x2_square = data['x2'] ** 2
x1_x2 = data['x1'] * data['x2']

# New DataFrame for these derived columns
derived_data = pd.DataFrame({'x1^2': x_square, 'x2^2': x2_square, 'x1*x2': x1_x2})
# Combine the original data with the new columns
data_combined = pd.concat([data, derived_data], axis=1)
```

### Why these transformations?
- **Polynomial Features** help model nonlinear relationships between features and the target variable.
- By adding **squared terms** (`x1_square` and `x2_square`) and interaction terms (`x1 * x2`), you're extending the original feature space, making it possible for algorithms like **Polynomial Regression** or **SVM with Polynomial Kernel** to learn more complex relationships.

These types of transformations are useful for creating decision boundaries that aren't linear (curved). When we apply these transformations, we’re expanding the feature space, making it possible for algorithms to find higher-order relationships and predict better.

### How it helps in Polynomial Kernel (for SVM)?
In SVM, when you use a **polynomial kernel**, you're implicitly mapping the data into a higher-dimensional space using a transformation like:

- **K(x, x') = (x ⋅ x' + c)^d**, where `c` is a constant, and `d` is the degree of the polynomial. 
- Polynomial features like `x1^2`, `x2^2`, and `x1*x2` are manually creating similar feature space transformations. 

So, your original features might appear linear in the original space, but after using polynomial features, it might not be linear anymore in the higher-dimensional space. The algorithm finds the optimal hyperplane for this space.

This expanded feature space enables us to model non-linear relationships more effectively.

### Visual Example
- If you just use `x1` and `x2` (linear data), the data would be scattered in a way that makes it hard for a linear model to separate classes.
- However, after adding `x1^2`, `x2^2`, and `x1 * x2`, the data becomes more "stacked" (in higher-dimensional space), and a polynomial decision surface can better separate the data.


```python
import plotly.express as px
# Plotting using Plotly
fig = px.scatter_3d(
    data_combined, 
    x='x1', 
    y='x2', 
    z='x1*x2',  # Using x1*x2 for z-axis
    color='y',  # Optional, to color the points based on the target variable
    title="3D Scatter Plot of x1, x2, and x1*x2",
    labels={"x1": "X1", "x2": "X2", "x1*x2": "X1*X2", "y": "Target (y)"}
)

# Show the plot
fig.show()
```

### Now plot , x1_square, x2_square and x1*x2 and you will se the data points are well seperated.

```python
import plotly.express as px
# Plotting using Plotly
fig = px.scatter_3d(
    data_combined, 
    x='x1_square', 
    y='x2_square', 
    z='x1*x2',  # Using x1*x2 for z-axis
    color='y',  # Optional, to color the points based on the target variable
    title="3D Scatter Plot of x1 square, x2 square, and x1*x2",
    labels={"x1": "X1", "x2": "X2", "x1*x2": "X1*X2", "y": "Target (y)"}
)

# Show the plot
fig.show()
```

### So from above we can see we are using the polynomial formula manually to perform on the features and seperate my datpaoint in the higher env. The same we can do using Poly kernel, which takes care internally

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

### Accuracy is very low when kernel is linear, as we can see our datapoints are Highly stacked and grouped with each other

```python
svc_classifier = SVC(kernel='linear')
svc_classifier.fit(X_train, Y_train)
y_pred = svc_classifier.predict(X_test)
print('Score: ',accuracy_score(Y_test, y_pred))
```

### Accuracy is less when we use poly

```python
svc_classifier = SVC(kernel='poly')
svc_classifier.fit(X_train, Y_train)
y_pred = svc_classifier.predict(X_test)
print('Score: ',accuracy_score(Y_test, y_pred))
```

### Accuracy is better when we use rbf

```python
svc_classifier = SVC(kernel='rbf')
svc_classifier.fit(X_train, Y_train)
y_pred = svc_classifier.predict(X_test)
print('Score: ',accuracy_score(Y_test, y_pred))
```

### Accuracy is very low when we use sigmoid

```python
svc_classifier = SVC(kernel='sigmoid')
svc_classifier.fit(X_train, Y_train)
y_pred = svc_classifier.predict(X_test)
print('Score: ',accuracy_score(Y_test, y_pred))
```
